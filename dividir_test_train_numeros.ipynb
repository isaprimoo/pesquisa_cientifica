COM ESSE CÓDIGO AQUI, ELE PEGA A PASTA QUE CONTEM PASTAS DIVIDIDAS DO 1 AO 9, E PEGA CADA UMA DAS PASTAS E DIVIDE EM 80% PARA TREINO E 20% PARA TESTE, DO PRÓPRIO NÚMERO.


OBS: ELE CRIA UM NOVO DIRETÓRIO QUE CONTEM OS DADOS DISTRIBUÍDOS EM TREINO/TESTE E DENTRO DE CADA UM DIRETORIO DE CADA UM NUMERO.


```python
import os
import random
import shutil

base_dir = "C:/Users/Downloads/REVISAO_NUMEROS_PESQUISA_CIENTIFICA/REVISAO_NUMEROS_PESQUISA_CIENTIFICA/REVISAO_NUMEROS_PESQUISA_CIENTIFICA"
saida_dir = os.path.join(base_dir, "DIVISAO_TREINO_TESTE")

for classe in range(0, 10):
    pasta_classe = os.path.join(base_dir, str(classe))
    if not os.path.isdir(pasta_classe):
        continue

    # Cria as pastas destino fora da pasta original
    train_dir = os.path.join(saida_dir, "TRAIN", str(classe))
    test_dir = os.path.join(saida_dir, "TEST", str(classe))
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # Lista apenas imagens da pasta original (ignora subpastas)
    imagens = [f for f in os.listdir(pasta_classe)
               if os.path.isfile(os.path.join(pasta_classe, f)) and f.lower().endswith('.png')]

    random.shuffle(imagens)

    n_test = int(0.2 * len(imagens))
    test_imgs = imagens[:n_test]
    train_imgs = imagens[n_test:]

    # Copia em vez de mover
    for img in train_imgs:
        shutil.copy(os.path.join(pasta_classe, img), os.path.join(train_dir, img))
    for img in test_imgs:
        shutil.copy(os.path.join(pasta_classe, img), os.path.join(test_dir, img))
```
