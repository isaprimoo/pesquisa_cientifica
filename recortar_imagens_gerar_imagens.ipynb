```python
from PIL import Image
import os
import glob # Usado para encontrar todos os arquivos com um padrão

# --- CONFIGURAÇÃO DE DIRETÓRIOS ---
# Diretório onde estão suas imagens originais (com 5 dígitos)
INPUT_DIR = r'C:\Users\x646688\OneDrive - rede.sp\Documentos\PESQUISA CIENTIFICA\SCUT-WMN DataSet\easy_samples\imagens'

# Diretório onde os dígitos recortados serão salvos.
# Ele será criado se não existir.
OUTPUT_DIR = r'C:\Users\x646688\OneDrive - rede.sp\Documentos\PESQUISA CIENTIFICA\SCUT-WMN DataSet\easy_samples\digitos_recortados'

# Cria o diretório de saída se ele não existir
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Padrão para encontrar todas as imagens PNG (você pode adicionar .jpg, etc., se necessário)
image_pattern = os.path.join(INPUT_DIR, '*.png')

# Itera sobre todos os arquivos que correspondem ao padrão no diretório de entrada
for image_path in glob.glob(image_pattern):
    try:
        # 1. CARREGAR, CONVERTER E BINARIZAR A IMAGEM
        # O nome do arquivo original (ex: '10438.png') é importante para nomear os dígitos de saída
        file_name = os.path.basename(image_path)
        base_name, ext = os.path.splitext(file_name) # '10438', '.png'

        print(f"Processando imagem: {file_name}")

        # Carregar imagem e converter para escala de cinza ("L")
        img = Image.open(image_path).convert("L")

        # Binário: p > 127 é 255 (branco), senão é 0 (preto)
        # O bbox original não está sendo usado no seu código de recorte,
        # mas pode ser útil para remover bordas brancas se fosse necessário.
        # binary = img.point(lambda p: 255 if p > 127 else 0)
        # bbox = binary.getbbox()

        # 2. CALCULAR DIMENSÕES E DIVIDIR EM 5 PARTES IGUAIS
        w, h = img.size

        # AQUI ELE DIVIDE A IMAGEM EM CINCO PARTES IGUAIS
        digit_width = w // 5

        # 3. RECORTAR E SALVAR OS 5 DÍGITOS
        for i in range(5):
            # Define a caixa de corte (esquerda, topo, direita, fundo)
            box = (i * digit_width, 0, (i + 1) * digit_width, h)

            # Recorta a porção da imagem
            digit = img.crop(box)

            # Define o nome do arquivo de saída:
            # Exemplo: '10438_digito_0.png', '10438_digito_1.png', etc.
            output_file_name = f"{base_name}_digito_{i}{ext}"
            output_path = os.path.join(OUTPUT_DIR, output_file_name)

            # Salva o dígito recortado no diretório de saída
            digit.save(output_path)

        print(f"   -> 5 dígitos salvos de {file_name} em {OUTPUT_DIR}")

    except Exception as e:
        print(f"Erro ao processar {image_path}: {e}")

print("\nProcessamento de todas as imagens concluído!")
```
